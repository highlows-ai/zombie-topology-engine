```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from src.load_dataset import load_topology_dataset
```

```python
df = load_topology_dataset()
df.head()
```

```python
topology_markers = {
    "Scarcity": "o",
    "Safety": "^",
    "Compliance": "s",
    "Infrastructure": "D",
    "None": "x"
}

lifecycle_opacity = {
    "Pre": 0.4,
    "Mid": 0.9,
    "Terminal": 0.6
}
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection='3d')

for _, row in df.iterrows():
    ax.scatter(
        row["ZombieState"],
        row["RegulatoryDensity"],
        row["CapitalIntensity"],
        c=row["ReturnProfile"],
        cmap="viridis",
        marker=topology_markers[row["RegulatoryTopology"]],
        alpha=lifecycle_opacity[row["LifecycleStage"]],
        s=80
    )

ax.set_xlabel("Zombie State")
ax.set_ylabel("Regulatory Density")
ax.set_zlabel("Capital Intensity")

plt.title("Full Industrial Topology — 3D Map")
plt.show()
```

```python
zombie_quadrant = [
    "Funeral Homes",
    "Storage Facilities",
    "Car Washes",
    "Laundromats",
    "ATM Networks",
    "Vending Routes",
    "Property Management",
    "Local Government Contracting",
    "Pest Control",
    "Elevator Maintenance",
    "Fire Safety Inspection",
    "Water Testing & Treatment",
    "HVAC Service Companies"
]

df_zq = df[df["Industry"].isin(zombie_quadrant)]
df_zq
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection='3d')

for _, row in df_zq.iterrows():
    ax.scatter(
        row["ZombieState"],
        row["RegulatoryDensity"],
        row["CapitalIntensity"],
        c=row["ReturnProfile"],
        cmap="plasma",
        marker=topology_markers[row["RegulatoryTopology"]],
        alpha=lifecycle_opacity[row["LifecycleStage"]],
        s=120
    )

ax.set_xlabel("Zombie State")
ax.set_ylabel("Regulatory Density")
ax.set_zlabel("Capital Intensity")

plt.title("Zombie-Return Quadrant — Allocator Gold")
plt.show()
```
